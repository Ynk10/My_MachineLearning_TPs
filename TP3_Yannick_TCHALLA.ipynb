# Exercice 1 :  entraîner et tester différents classifieurs sur un jeu de données factice



```python
#1-Créer un jeu de données factice :

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import mglearn

X, y = make_blobs(n_samples=30, centers=2, random_state=42)
```

```python
#2-Tracer les lignes de décision pour deux classifieurs :
        #Créons une figure avec deux sous-graphiques.
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

#3-Ajuster les modèles sur le jeu de données :

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

logreg = LogisticRegression().fit(X, y)
svm = LinearSVC().fit(X, y)

#4-Tracer les lignes de décision pour les 2 modèles :

mglearn.plots.plot_2d_separator(logreg, X, eps=0.5, ax=axes[0], alpha=.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=axes[0])
axes[0].set_title("Logistic Regression")

mglearn.plots.plot_2d_separator(svm, X, eps=0.5, ax=axes[1], alpha=.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=axes[1])
axes[1].set_title("Linear SVC")

plt.show()
```

```python
#5-Afficher les scores de classification pour les deux classifieurs :
print("Logistic Regression score:", logreg.score(X, y))
print("Linear SVC score:", svm.score(X, y))
```

6-

```python
#7-Changer la valeur de pénalité C pour le classifieur SVC

fig, axes = plt.subplots(1, 3, figsize=(15, 3))

for ax, C in zip(axes, [0.01, 1.0, 100.0]):
    svm = LinearSVC(C=C).fit(X, y)
    mglearn.plots.plot_2d_separator(svm, X, eps=0.5, ax=ax, alpha=.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("Linear SVC (C={})".format(C))

plt.show()
```

# Exercice 2 :  entraîner et tester sur un jeu de données factice avec changement de paramètres

```python
#1-Créer un jeu de données factice avec centers=4

X, y = make_blobs(n_samples=30, centers=4, random_state=42)
y = y % 2
```

```python
# Tracer le nuage de points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Nuage de points avec 2 classes")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Classe binaire")
plt.show()
```

```python
#2-Entrainer un SVC linéaire, un polynomiale et un gaussien (rbf) :

from sklearn.svm import SVC

linear_svm = LinearSVC().fit(X, y)
poly_svm = SVC(kernel='poly').fit(X, y)
rbf_svm = SVC(kernel='rbf').fit(X, y)
```

```python
#3-Comparer les scores d'entraînement et expliquer les résultats :

print("Linear SVC score:", linear_svm.score(X, y))
print("Polynomial SVC score:", poly_svm.score(X, y))
print("RBF SVC score:", rbf_svm.score(X, y))
```

Explication du resultat: Le score des approches RBF et polynomial est plus élevé car elles peuvent apprendre des frontières de décision plus complexes que l'approche linéaire

# Exercice 3 : un ensemble de données factice à 3 classes

```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

#1-Créer un ensemble de données factice à 3 classes :
X, y = make_blobs(n_samples=90, centers=3, random_state=42)

#2-Ré-entraîner le modèle sur les nouvelles données :
logreg = LogisticRegression().fit(X, y)
svm = LinearSVC().fit(X, y)
```

Il devrait y avoir trois lignes de décision dérivées pour chaque modèle. Les coefficients et les ordonnées à l'origine des lignes de décision peuvent être utilisés pour tracer ces lignes.

```python

#3-Trouver les caractéristiques des lignes de décision :
print("Logistic Regression coefficients:", logreg.coef_)
print("Logistic Regression intercepts:", logreg.intercept_)

print("Linear SVC coefficients:", svm.coef_)
print("Linear SVC intercepts:", svm.intercept_)
```

```python

#4-Tracer les points de l'ensemble de données avec les lignes de décision :
plt.figure(figsize=(10, 6))
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

# Tracer les lignes de décision pour la régression logistique
coef = logreg.coef_
intercept = logreg.intercept_

for i in range(len(coef)):
    coef_i = coef[i]
    intercept_i = intercept[i]
    x_values = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100)
    y_values = -(coef_i[0] * x_values + intercept_i) / coef_i[1]
    plt.plot(x_values, y_values, label=f"Logistic Regression Decision Boundary {i}", linestyle='--')

# Tracer les lignes de décision pour le SVC linéaire
coef = svm.coef_
intercept = svm.intercept_

for i in range(len(coef)):
    coef_i = coef[i]
    intercept_i = intercept[i]
    x_values = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100)
    y_values = -(coef_i[0] * x_values + intercept_i) / coef_i[1]
    plt.plot(x_values, y_values, label=f"Linear SVC Decision Boundary {i}", linestyle='-.')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Decision Boundaries for Logistic Regression and Linear SVC")
plt.legend()
plt.show()

```


# Exercice 4 : Arbres de décision

```python
#1-Importer la classe DecisionTreeClassifier et charger un jeu de données  : 

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Charger un jeu de données pour la classification
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
#2-Initialiser un objet DecisionTreeClassifier et ajuster le modèle :

tree_clf = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_clf.fit(X_train, y_train)
```

```python
#3-Évaluer les performances du modèle :

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Prédire les étiquettes pour les données de test
y_pred = tree_clf.predict(X_test)

# Calculer différentes mesures de performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
#4-Explorer l'importance des caractéristiques :

print("Feature Importances:", tree_clf.feature_importances_)
```

```python
#5-Visualiser la structure de l'arbre de décision :

from sklearn.tree import plot_tree

plt.figure(figsize=(12, 8))
plot_tree(tree_clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```


# Exercice 5 : Forêts aléatoires

```python
#1-Importer la classe RandomForestClassifier et charger un jeu de données :

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Charger un jeu de données pour la classification
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
#2-Initialiser un objet RandomForestClassifier et ajuster le modèle :

forest_clf = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42)
forest_clf.fit(X_train, y_train)
```

```python
#3-Évaluer les performances du modèle :

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Prédire les étiquettes pour les données de test
y_pred = forest_clf.predict(X_test)

# Calculer différentes mesures de performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
#4-Explorer l'importance des caractéristiques :

print("Feature Importances:", forest_clf.feature_importances_)
```

```python
#5-Visualiser la frontière de décision de la forêt aléatoire en 2D :

import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

# Vérifier si les données ont moins de 3 caractéristiques pour pouvoir les visualiser en 2D
if X_train.shape[1] <= 2:
    plt.figure(figsize=(10, 6))
    plot_decision_regions(X_train, y_train, clf=forest_clf, legend=2)
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])
    plt.title("Decision Boundary of Random Forest Classifier")
    plt.show()
else:
    print("Les données ont plus de 2 caractéristiques, la visualisation en 2D n'est pas possible.")
```
